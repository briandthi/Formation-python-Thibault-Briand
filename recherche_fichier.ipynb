```python
class WrongFileTypeError(Exception):
    
    def __init__ ( self , message = "") :
        super().__init__(message)
        
class FileSizeError(Exception):
    
    def __init__ ( self , message = "") :
        super().__init__(message)

```

```python
def recc_folder_search(folder_path, size = 1000000000000, end = ''):
    
    list_doss = {}
    nb_py = 0
    nb_filtres = 0
    nb_fichier = 0
    
    for elmt in os.listdir(folder_path):
        path = folder_path + '/' + elmt
        if os.path.isdir(f"{folder_path}/{elmt}"):
            res = recc_folder_search(f"{folder_path}/{elmt}", size, end)
            nb_py += res[0]
            nb_filtres += res[1]
            nb_fichier += res[2]
            list_doss[elmt] = res[3]
            
        else:
            nb_fichier += 1
            
            if os.stat(path).st_size < size:
                if path.endswith(end):
                    nb_py += 1
                    list_doss[elmt] = f'{os.stat(f"{folder_path}/{elmt}").st_size} octets'
                else :
                    nb_filtres += 1
                    # print(WrongFileTypeError(f'{elmt} is not a {end} file'))
            else:
                nb_filtres += 1
                # print(FileExistsError( f'{elmt} is too big: {os.stat(f"{folder_path}/{elmt}").st_size} octet'))
            
    if list_doss=={}:
        return nb_py, nb_filtres, nb_fichier, "EMPTY FOLDER"
    
    else:
        return nb_py, nb_filtres, nb_fichier, list_doss

res = recc_folder_search(os.curdir, 1000, '.py')

print(f"###### RESULTAT ########\nnombres de fichiers python: {res[0]}\nnombres de fichiers filtrés: {res[1]}\nnombres de fichiers total: {res[2]}\n")
print("Mon dossier:\n",json.dumps(res[3], indent = 4))
```
